```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 10)

print("Libraries imported successfully.")
```

```python
# Load dataset
df = pd.read_excel("../data/raw/Online Retail.xlsx")

print("Dataset loaded successfully.")
df.head()
```

```python
# Dataset shape
df.shape
```

```python
# Column names
df.columns
```

```python
# Data types
df.info()
```

```python
# Missing values
df.isnull().sum()
```

```python
# Duplicate rows
df.duplicated().sum()
```

```python
# Remove duplicate rows
df = df.drop_duplicates()

print("New shape after removing duplicates:", df.shape)
```

```python
# Missing values in Description
df["Description"].isnull().sum()
```

```python
# Remove rows with missing Description
df = df.dropna(subset=["Description"])

print("New shape after removing missing descriptions:", df.shape)
```

```python
# Invoices that start with "C" (cancelled/returned orders)
df[df["InvoiceNo"].astype(str).str.startswith("C")].shape
```

```python
# Rows with negative quantity
df[df["Quantity"] < 0].shape
```

```python
# Rows with zero quantity
df[df["Quantity"] == 0].shape
```

```python
# Rows with invalid UnitPrice
df[df["UnitPrice"] <= 0].shape
```

```python
# Keep only valid sales transactions
df = df[
    (df["Quantity"] > 0) &
    (df["UnitPrice"] > 0) &
    (~df["InvoiceNo"].astype(str).str.startswith("C"))
]

print("Shape after removing returns/cancellations and invalid values:", df.shape)
```

```python
# Create TotalSales column
df["TotalSales"] = df["Quantity"] * df["UnitPrice"]

df.head()
```

```python
# Create date-related columns
df["Year"] = df["InvoiceDate"].dt.year
df["Month"] = df["InvoiceDate"].dt.month
df["MonthName"] = df["InvoiceDate"].dt.strftime("%B")
df["MonthYear"] = df["InvoiceDate"].dt.to_period("M").astype(str)

df.head()
```

```python
# Check updated dataframe structure
df.info()
```

```python
# Preview cleaned and enriched dataset
df.head()
```

```python
# Export cleaned dataset
df.to_csv("../data/processed/online_retail_cleaned.csv", index=False)

print("Cleaned dataset exported successfully.")
```

```python
# Create monthly summary by product
monthly_summary = df.groupby(["MonthYear", "Description"], as_index=False).agg({
    "Quantity": "sum",
    "UnitPrice": "mean",
    "TotalSales": "sum"
})

monthly_summary.rename(columns={
    "Quantity": "TotalQuantitySold",
    "UnitPrice": "AverageUnitPrice"
}, inplace=True)

monthly_summary.head()
```

```python
# Export monthly summary
monthly_summary.to_csv("../data/processed/monthly_summary.csv", index=False)

print("Monthly summary exported successfully.")
```

```python
# Rebuild monthly summary cleanly
monthly_summary = df.groupby(["MonthYear", "Description"], as_index=False).agg({
    "Quantity": "sum",
    "UnitPrice": "mean",
    "TotalSales": "sum"
})

monthly_summary.rename(columns={
    "Quantity": "TotalQuantitySold",
    "UnitPrice": "AverageUnitPrice"
}, inplace=True)

# Force numeric formatting
monthly_summary["AverageUnitPrice"] = monthly_summary["AverageUnitPrice"].astype(float).round(2)
monthly_summary["TotalSales"] = monthly_summary["TotalSales"].astype(float).round(2)

monthly_summary.head(10)
```
